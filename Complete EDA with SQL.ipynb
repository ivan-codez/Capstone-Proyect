```python
%sql create table SPACEXTABLE as select * from SPACEXTBL where Date is not null


```

## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
import pandas as pd

df_sql = pd.read_sql_query("SELECT * FROM SPACEXTABLE LIMIT 5", con)
df_sql





```

```python
pd.read_sql_query(
    "SELECT DISTINCT Launch_Site FROM SPACEXTABLE",
    con
)
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'KSC' 


```python
pd.read_sql_query(
    "SELECT * FROM SPACEXTABLE WHERE Launch_Site LIKE 'KSC%' LIMIT 5",
    con
)
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
pd.read_sql_query(
    """
    SELECT SUM(PAYLOAD_MASS__KG_) AS Total_Payload_Mass
    FROM SPACEXTABLE
    WHERE Customer LIKE '%NASA%'
      AND Customer LIKE '%CRS%'
    """,
    con
)
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
pd.read_sql_query(
    """
    SELECT AVG("PAYLOAD_MASS__KG_") AS Avg_Payload_Mass
    FROM SPACEXTABLE
    WHERE Booster_Version = 'F9 v1.1'
    """,
    con
)

```

### Task 5

##### List the date where the succesful landing outcome in drone ship was acheived.


_Hint:Use min function_ 


```python
pd.read_sql_query(
    """
    SELECT MIN(Date) AS First_Successful_Drone_Landing
    FROM SPACEXTABLE
    WHERE Landing_Outcome LIKE 'Success%'
      AND Landing_Outcome LIKE '%drone%'
    """,
    con
)
```

### Task 6

##### List the names of the boosters which have success in ground pad  and have payload mass greater than 4000 but less than 6000


```python
pd.read_sql_query(
    """
    SELECT DISTINCT Booster_Version
    FROM SPACEXTABLE
    WHERE Landing_Outcome LIKE 'Success%'
      AND Landing_Outcome LIKE '%ground%'
      AND "PAYLOAD_MASS__KG_" BETWEEN 4000 AND 6000
    """,
    con
)
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
pd.read_sql_query(
    """
    SELECT Mission_Outcome, COUNT(*) AS Total
    FROM SPACEXTABLE
    GROUP BY Mission_Outcome
    """,
    con
)
```

### Task 8



##### List all the booster_versions that have carried the maximum payload mass. Use a subquery.


```python
pd.read_sql_query(
    """
    SELECT DISTINCT Booster_Version
    FROM SPACEXTABLE
    WHERE "PAYLOAD_MASS__KG_" = (
        SELECT MAX("PAYLOAD_MASS__KG_")
        FROM SPACEXTABLE
    )
    """,
    con
)

```

### Task 9


##### List the records which will display the month names, succesful landing_outcomes in ground pad ,booster versions, launch_site for the months in year 2017
**Note: SQLLite does not support monthnames. So you need to use substr(Date,6,2) for month, substr(Date,9,2) for date, substr(Date,0,5),='2017' for year.**


```python
pd.read_sql_query(
    """
    SELECT 
        substr(Date,6,2) AS Month,
        Landing_Outcome,
        Booster_Version,
        Launch_Site
    FROM SPACEXTABLE
    WHERE substr(Date,1,4) = '2017'
      AND Landing_Outcome LIKE 'Success%'
      AND Landing_Outcome LIKE '%ground%'
    """,
    con
)
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
pd.read_sql_query(
    """
    SELECT 
        Landing_Outcome,
        COUNT(*) AS Total
    FROM SPACEXTABLE
    WHERE Date BETWEEN '2010-06-04' AND '2017-03-20'
    GROUP BY Landing_Outcome
    ORDER BY Total DESC
    """,
    con
)
```
